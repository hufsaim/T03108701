## Markov matrix
N명의 학생의 홈페이지가 있고, 각 학생의 홈페이지에는 다른 학생들의 홈페이지로 이동할 수 있는 링크를 포함하고 있다고 하자.

예를 들어 3명의 학생 철수, 민수, 영희가 있을 때에
- 철수의 홈페이지에는 철수, 민수에 대한 링크가 있고,
- 민수의 홈페이지에는 철수, 영희에 대한 링크가 있고,
- 영희의 홈페이지에는 철수, 민수, 영희에 대한 링크가 있다고 하자.

각 matrix의 column을 해당 학생의 홈페이지에서 다른 학생의 홈페이지로 이동할 수 있는 확률이라고 하고, 각 링크로 이동할 확률은 균일하다고 하면 우리는 해당정보를 요약하는 다음과 같은 3 x 3 matrix A를 정의할 수 있다.

$A =
 \begin{pmatrix}
  1/2 & 1/2 & 1/3 \\
  1/2 & 0  & 1/3 \\
  0 & 1/2 & 1/3
 \end{pmatrix}$

각 column, row는 순서대로 철수, 민수, 영희를 나타낼 때, $A^2$, $A^3$, ... , $A^{100}$ 이 어떤 정보를 주는 지 생각하여 보자.

```python
A = np.array([[1/2, 1/2, 1/3],
              [1/2, 0, 1/3],
              [0, 1/2, 1/3]])
print(A)
```

```python
print(np.linalg.matrix_power(A,2))
```

```python
print(np.linalg.matrix_power(A,3))
```

```python
print(np.linalg.matrix_power(A,5))
```

```python
A100 = np.linalg.matrix_power(A,100)
print(A100)
```

## Eigenvalues, eigenvectors of Markov matrix
위에서 정의한 matrix A의 eigenvalue와 eigenvector를 구하여 $A^{100}$의 결과와 함께 이해하여 보자.

```python
eigvals, eigvecs = np.linalg.eig(A)
print(eigvals)
print(eigvecs)
```

```python
eigvecs1 = eigvecs[:,np.argmax(eigvals)]
print(eigvecs1)
```

```python
r = A100[:,1]/eigvecs1
```

```python
print(eigvecs1 * (r[0]))
```

```python
print(A100[:,0])
```
