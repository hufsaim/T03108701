```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 4
A = np.zeros((N,N))
for k in range(N):
  for n in range(N):
    A[k,n] = np.cos( (np.pi/N) * (n + 0.5) * k )
  plt.figure(figsize=(5,1))
  plt.plot(A[k,:])
  plt.title('row ' + str(k))
  plt.show()
print(A)
```

```python
N = 32
A = np.zeros((N,N))
for n in range(N):
  for k in range(N):
    A[k,n] = np.cos( (np.pi/N) * (n + 0.5) * k )

plt.imshow(A,cmap='jet')
plt.colorbar()
plt.show()
```

```python
plt.imshow(A.transpose(),cmap='jet')
plt.colorbar()
plt.show()
```

```python
ATA = np.matmul(A.transpose(),A)
plt.imshow(ATA,cmap='jet')
plt.colorbar()
plt.show()
```
