```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('MBTI2024.xlsx')
print(df.shape)
df.columns[1:]
```

## 2D vectors

```python
s1 = '홍길동'
s2 = '홍길서'
s3 = '홍길남'

v1 = df[s1].to_numpy()[[3,5]]
v2 = df[s2].to_numpy()[[3,5]]
v3 = df[s3].to_numpy()[[3,5]]

print(v1)
print(v2)
print(v3)
```

## Visualization of 2D vectors

```python
plt.figure(dpi=128)
ax=plt.gca()

ax.set_aspect('equal')
ax.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='r')
#ax.text(v1[0],v1[1],s1[1:])
ax.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='g')
#ax.text(v2[0],v2[1],s2[1:])
ax.quiver(0,0,v3[0],v3[1],angles='xy',scale_units='xy',scale=1,color='b')
#ax.text(v3[0],v3[1],s3[1:])

ax.set_xlim([-1.05,1.05])
ax.set_ylim([-1.05,1.05])
ax.set_xlabel('E-I')
ax.set_ylabel('T-F')
ax.grid('both')
```

## Dot product of two 2D vectors

```python
print(f'{s1} and {s2}: {np.dot(v1,v2)}')
print(f'{s2} and {s3}: {np.dot(v2,v3)}')
print(f'{s3} and {s1}: {np.dot(v3,v1)}')
```

## Visualization of all 2D vectors by points

```python
plt.figure(dpi=128)
ax=plt.gca()
ax.set_aspect('equal')
ax.scatter(0,0,s=5,color='g')
for n in np.arange(1,df.shape[-1]):
    if df.iloc[[2],n].values[0]==1:
        ax.scatter(df.iloc[[3],n],df.iloc[[5],n],s=3,color='b')
    else:
        ax.scatter(df.iloc[[3],n],df.iloc[[5],n],s=3,color='r')
   # ax.text(df.iloc[[3],n],df.iloc[[5],n],df.columns[n][1:])
ax.grid('on')
ax.set_xlim([-1.05,1.05])
ax.set_ylim([-1.05,1.05])
ax.set_xlabel('E-I')
ax.set_ylabel('T-F')
ax.grid('both')
```

## Dot product of two 5D vectors

```python
v1 = df[s1].to_numpy()[[3,4,5,6,7]]
v2 = df[s2].to_numpy()[[3,4,5,6,7]]
v3 = df[s3].to_numpy()[[3,4,5,6,7]]
print(s1,v1)
print(s2,v2)
print(s3,v3)
```

```python
print(f'{s1} and {s2}: {np.dot(v1,v2)}')
print(f'{s2} and {s3}: {np.dot(v2,v3)}')
print(f'{s3} and {s1}: {np.dot(v3,v1)}')
```
